```python
!pip install reportlab matplotlib pandas
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet
import matplotlib.pyplot as plt
import pandas as pd
pdf_path = "/content/my_AI_project_report.pdf"
doc = SimpleDocTemplate(pdf_path, pagesize=A4)
styles = getSampleStyleSheet()
content = []
title = Paragraph("<b><font size=16>AI Project Report</font></b>", styles["Title"])
intro = Paragraph("This report was created by <b>Reshma ðŸŒ¸</b> using Python in Google Colab. It includes text, images, tables, and charts.", styles["BodyText"])
content.extend([title, Spacer(1, 12), intro, Spacer(1, 20)])
data = [
    ["Model", "Accuracy", "Dataset"],
    ["AI Model 1", "92%", "Math Quiz Data"],
    ["AI Model 2", "88%", "English Worksheet Data"],
]

table = Table(data)
table.setStyle(TableStyle([
    ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
    ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
    ("ALIGN", (0, 0), (-1, -1), "CENTER"),
    ("GRID", (0, 0), (-1, -1), 1, colors.black)
]))
content.extend([Paragraph("<b>Project Results Table:</b>", styles["Heading2"]), table, Spacer(1, 20)])
# Create chart
scores = [92, 88, 85]
models = ["Model 1", "Model 2", "Model 3"]

plt.figure(figsize=(4,3))
plt.bar(models, scores)
plt.title("Model Performance")
plt.savefig("/content/chart.png")

# Add chart to PDF
chart = Image("/content/chart.png", width=300, height=200)
content.extend([Paragraph("<b>Performance Chart:</b>", styles["Heading2"]), chart, Spacer(1, 20)])
image_path = "/content/sample_image.jpg"  # change to your uploaded image name
img = Image(image_path, width=200, height=150)
content.extend([img, Spacer(1, 20)])
doc.build(content)
print("âœ… PDF created successfully:", pdf_path)
from google.colab import files
files.download(pdf_path)
```
